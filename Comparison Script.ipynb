```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import umap
from sklearn.datasets import load_iris

# Load the Iris dataset
data = load_iris()
X = data.data
y = data.target

# Apply PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Apply t-SNE
tsne = TSNE(n_components=2, perplexity=30, random_state=42)
X_tsne = tsne.fit_transform(X)

# Apply UMAP
umap_model = umap.UMAP(n_components=2, random_state=42)
X_umap = umap_model.fit_transform(X)

# Plot the results
fig, ax = plt.subplots(1, 3, figsize=(18, 6))

# Plot PCA
ax[0].scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis', edgecolor='k', s=150)
ax[0].set_title("PCA")
ax[0].set_xlabel('Component 1')
ax[0].set_ylabel('Component 2')

# Plot t-SNE
ax[1].scatter(X_tsne[:, 0], X_tsne[:, 1], c=y, cmap='viridis', edgecolor='k', s=150)
ax[1].set_title("t-SNE")
ax[1].set_xlabel('Component 1')
ax[1].set_ylabel('Component 2')

# Plot UMAP
ax[2].scatter(X_umap[:, 0], X_umap[:, 1], c=y, cmap='viridis', edgecolor='k', s=150)
ax[2].set_title("UMAP")
ax[2].set_xlabel('Component 1')
ax[2].set_ylabel('Component 2')

plt.show()
```
